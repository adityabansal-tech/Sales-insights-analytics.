# Sales Insights — Exploratory Analysis
Quick ad-hoc queries against the loaded database, outside the main KPI views.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import pandas as pd
from sales_engine import load

engine = load.get_engine()
```

```python
# Monthly revenue trend
df = pd.read_sql('SELECT * FROM vw_monthly_revenue', engine)
df.plot(x='month_name', y='total_revenue', kind='bar', figsize=(12,4), title='Monthly Revenue')
```

```python
# Top products by profit margin
df = pd.read_sql('SELECT * FROM vw_product_performance', engine)
df['margin_pct'] = (df['total_profit'] / df['total_revenue'] * 100).round(2)
df.sort_values('margin_pct', ascending=False).head(10)
```

```python
# Category share of total revenue
df = pd.read_sql('SELECT * FROM vw_category_performance', engine)
df.plot.pie(y='total_revenue', labels=df['category'], figsize=(6,6), autopct='%1.1f%%', legend=False, title='Revenue Share by Category')
```
